```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Gerando dados sintéticos
np.random.seed(42)  # Garantindo a reproducibilidade

X = 2 * np.random.rand(100, 1)  # 100 amostras de nivel de escolaridade
y = 4 + 3 * X**2 + np.random.randn(100, 1)  # y = 4 + 3X^2 + ruído


# Step 2: Add bias term (intercept) to X
X = np.c_[np.ones((100, 1)), X]  # Adicionando x0 = 1 (intercept ou bias)

# Step 3: Inicialização dos parâmetros (weights)
theta = np.random.randn(2, 1)  # 2 parametros a serem estimados: theta_0 (bias) and theta_1 (slope)
theta

# Step 4: Definição dos Hiperparâmetros
learning_rate = 0.01  # Step size para updates
n_iterations = 1000 # Número de épocas

# Step 5: Algoritmo do Gradiente descendente
m = len(X)  # Número de amostras
cost_history = []

for iteration in range(n_iterations):
    gradients = (2/m) * X.T.dot(X.dot(theta) - y)  # Calcula os gradientes (derivadas)
    theta -= learning_rate * gradients  # Atualiza os pesos usando o gradiente descendente

    # Calcula o Mean Squared Error (MSE)
    mse = np.mean((X.dot(theta) - y) ** 2)
    cost_history.append(mse)

# Step 6: Gráficos da função de custo
plt.plot(range(n_iterations), cost_history)
plt.xlabel("Iterations")
plt.ylabel("Mean Squared Error")
plt.title("Cost Function Convergence")
plt.show()

# Step 7: Função final aprendida
plt.scatter(X[:,1], y, label="Data points")
plt.plot(X[:,1], X.dot(theta), color="red", label="Regression line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression Fit using Gradient Descent")
plt.show()

# Step 8: Mostra parâmetros finais
print(f"Final Parameters: Intercept = {theta[0][0]:.2f}, Slope = {theta[1][0]:.2f}")
```

# Multi Layer Perceptron

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

# Gerando dados sintéticos
np.random.seed(42)
X = 2 * np.random.rand(100, 1)  # 100 amostras de nivel de escolaridade
y = 3 * X**2 + np.random.randn(100, 1)  # y = 3X^2 + ruído

# Normalização dos dados
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

y = y.reshape(-1, 1)  # Garantindo que y tenha a forma correta

# Construção do modelo MLP
model = keras.Sequential([
    keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(1)  # Camada de saída para regressão
])

# Compilação do modelo
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.01), loss='mse')

# Treinamento do modelo
history = model.fit(X_scaled, y, epochs=100, verbose=1)

# Plot da função de custo
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel("Mean Squared Error")
plt.title("Cost Function Convergence")
plt.show()

# Predição e visualização
X_test = np.linspace(0, 2, 100).reshape(-1, 1)
X_test_scaled = scaler.transform(X_test)
y_pred = model.predict(X_test_scaled)

plt.scatter(X, y, label="Data points")
plt.plot(X_test, y_pred, color="red", label="MLP Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("MLP Regression Fit")
plt.show()

# Exibição dos pesos aprendidos
weights, bias = model.layers[0].get_weights()
print(f"Final Parameters: Weights = {weights.flatten()}, Bias = {bias}")
```

1 - Utilizando o framework Keras, crie uma Rede Neural Perceptron de Multiplas Camadas (MLP) e execute o treinamento da mesma com a base de dados MNIST. Siga os passos sugeridos abaixo:


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train[0]
```

```python
y_train[0]
```

```python
# Os valores dos pixels variam de [0,255]. Normalize os valores dos pixels no intervalo [0,1]
x_train = x_train / 255
x_test = x_test / 255
```

```python
# Faça um reshape das images de 28x28 para 1x784
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)
```

```python
x_train.shape
```

```python
y_train
```

```python
# Converta os labels (y) usando o one-hot encoding (utilize  keras.utils.to_categorical para isso)
y_train = keras.utils.to_categorical(y_train, num_classes=10)
y_test = keras.utils.to_categorical(y_test, num_classes=10)
```

```python
# Construa o modelo MLP (parametrize as camadas da forma que achar melhor) referência: https://keras.io/guides/sequential_model/
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(784,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
# model.Sequential()
# model.add(layers.Dense(64, activation='relu', input_shape=(784,)))
# model.add(layers.Dense(32, activation='relu'))
# model.add(layers.Dense(10, activation='softmax'))
```

```python
# utilize o model.summary para verificar se o modelo está parametrizado corretamente.
model.summary()
```

```python
# Compile o modelo com os seguintes parâmetros optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Treine o modelo
callback = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5)
history = model.fit(x_train, y_train, epochs=20, batch_size=32, validation_split=0.2, callbacks=[callback])
```

```python
# Avalie o modelo usando mode.evaluate
model.evaluate(x_test, y_test)
```

```python
# Utilize o modelo treinado para fazer predições na base de teste
predictions = model.predict(x_test)
```

```python
# utilize a função abaixo para plotar as imagens de teste e visualizar as predições do modelo para ela.
def plot_image(i, predictions_array, true_label, img):
    true_label, img = np.argmax(true_label), img.reshape(28,28)
    print(predictions_array)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel(f"{predicted_label} ({true_label})", color=color)

# Plot some sample images with predictions
num_rows = 3
num_cols = 3
num_images = num_rows * num_cols
plt.figure(figsize=(2*num_cols, 2*num_rows))
for i in range(num_images):
    plt.subplot(num_rows, num_cols, i+1)
    plot_image(i, predictions[i], y_test[i], x_test[i])
plt.tight_layout()
plt.show()
```

```python
y_preds = np.argmax(predictions, axis=1)
y_target = np.argmax(y_test, axis=1)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_target, y_preds)

cm
```
